```python
import pandas as pd
df = pd.read_excel(r'dummy_ojol_transactions_fix.xlsx')
df
```

```python
df.dtypes
```

```python
from datetime import datetime

df['tahun_lahir'] = df['customer_birthdate'].dt.strftime('%Y')
df
```

```python
now = datetime.now()

now.year
for index, row in df.iterrows():
    df.loc[index, 'umur']=now.year - int (row['tahun_lahir'])
df
```

```python
for index, row in df.iterrows():
    if row['umur'] >=26:
        df.loc[index, 'ket_umur'] = 'Dewasa'
    else:
        df.loc[index, 'ket_umur'] = 'Remaja'
df
```

```python
for index, row in df.iterrows():
    if row['distance'] >=6:
        df.loc[index, 'jarak'] = 'Jauh'
    if row['distance'] >=3:
        df.loc[index, 'jarak'] = 'Sedang'
    else:
        df.loc[index, 'jarak'] = 'Dekat'
df
```

```python
df['date_order'] = df['date_order'].str.replace(".",":")
df
```

```python
df['date_finished'] = df['date_finished'].str.replace(".",":")
df
```

```python
for index, row in df.iterrows():
    df.loc[index, 'waktu_selesai'] = row['date_finished'][-5:]
df
```

```python
for index, row in df.iterrows():
    df.loc[index, 'waktu_order'] = row['date_order'][-5:]
df
```

```python
for index, row in df.iterrows():
    df.loc[index, 'biaya_delivery_perkm'] = row['amount_delivery']/row['distance']
df
```
